```python
import numpy as np
from matplotlib import pyplot as plt
```

### 4.1

#### a

```python
def omega_min_one(a):
    Omega_k = .7
    Omega_r = 5e-5
    Omega_m = .3

    return -Omega_k/(Omega_m*a**-1 + Omega_r*a**-2 + Omega_k)
```

```python
a_arr = np.linspace(1e-7,1,500)

plt.plot(a_arr,omega_min_one(a_arr), color='darkred',linewidth=2)
plt.xlabel("a")
plt.ylabel("1-$\Omega$")
```

```python
omega_min_one(1e-32)
```

#### b

```python
a_e = 1e-27
post_infl_a_arr = np.logspace(-27,0,500)
plt.plot(post_infl_a_arr, -1*omega_min_one(post_infl_a_arr), color='darkred', linewidth=2)

pre_infl_a_arr = a_e * np.exp(-1*np.arange(61))
pre_infl_omega_min_one = omega_min_one(a_e) * np.exp(2*np.arange(61))
plt.plot(pre_infl_a_arr, -1*pre_infl_omega_min_one, color='darkred', linewidth=2)

plt.xscale('log')
plt.yscale('log')
plt.ylabel("$\Omega$-1")
plt.xlabel('a')
```
